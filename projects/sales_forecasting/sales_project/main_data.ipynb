```python
def apply_rolling(df,col,window):
    rolling_result= df[col].rolling(window=window).mean()
    rolling_result = rolling_result.bfill()
    return rolling_result
    
```

## with agg function

```python
def apply_rolling_agg(df,col,window):
    
    rolling_result= df[col].rolling(window=window).agg(["mean", "std"])
    
    rolling_result = rolling_result.bfill()
    
    return rolling_result
    
```

```python
rolling_agg=apply_rolling_agg(data_ob1removed,'OB1_AYLIK_ORTALAMA',3)
rolling_agg
```

```python
rolling_agg_ad=apply_rolling_agg(data_ob_ad,'OB1_AYLIK_ORTALAMA',3)
rolling_agg_ad
```

```python
rolling_agg_kg=apply_rolling_agg(data_ob_kg,'OB1_AYLIK_ORTALAMA',3)
rolling_agg_kg
```
